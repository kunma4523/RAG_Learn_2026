## What is RAG?

RAG (Retrieval-Augmented Generation) combines:
1. **Retrieval**: Find relevant documents
2. **Augmentation**: Add context to prompt
3. **Generation**: Generate answer with LLM

```python
import sys
sys.path.insert(0, '..')

from src.retrievers.dense import SentenceTransformerRetriever
from src.generators.llm import OpenAIGenerator
from src.pipelines.standard import StandardRAGPipeline
```

## Create Sample Documents

```python
# Sample documents about RAG
documents = [
    """Retrieval-Augmented Generation (RAG) is a technique that enhances
    Large Language Models by retrieving relevant information from external
    knowledge bases. This helps reduce hallucinations and provides up-to-date
    information.""",
    
    """The three main components of RAG are:
    1. Retriever - finds relevant documents
    2. Reranker - improves relevance ordering
    3. Generator - produces the final answer""",
    
    """BM25 is a ranking function used in information retrieval.
    It scores documents based on term frequency and inverse document frequency.
    BM25 is a sparse retrieval method.""",
    
    """Dense retrieval uses neural network embeddings to find similar documents.
    Unlike sparse methods like BM25, dense retrieval captures semantic similarity.
    Examples include DPR and Contriever.""",
    
    """Self-RAG is a framework that allows the LLM to decide when to retrieve
    information and how to use it. It combines retrieval with self-reflection."""
]

print(f"Created {len(documents)} sample documents")
```

## Initialize RAG Pipeline

```python
# Initialize retriever
retriever = SentenceTransformerRetriever(
    model_name="BAAI/bge-base-zh-v1.5",
    top_k=3
)

# Index documents
retriever.index(documents)
print("Documents indexed successfully!")
```

## Test Retrieval

```python
# Test retrieval
query = "What is RAG and what are its components?"
results = retriever.retrieve(query)

print(f"Query: {query}\n")
print("Retrieved documents:")
for i, result in enumerate(results, 1):
    print(f"\n{i}. Score: {result.score:.4f}")
    print(f"   {result.text[:100]}...")
```
